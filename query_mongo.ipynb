# Querying through Mongo DB

```python
from pymongo import MongoClient
import pandas as pd
```

### Calling the service MongoDB

```python
client = MongoClient("localhost:27017")
```

```python
db = client["Ironhack"]
```

```python
comp = db.get_collection("companies")
```

### Regex for names, categories, tags, description

- Conditions

```python
companies_name = comp.find({'name':
                               {"$regex": 
                                '^(design|ux|ui|design|frontend|backend|gaming|game|gamer|web|develop|software)',
                                "$options" :'i'}})
```

```python
companies_cat = comp.find({'category_code':
                               {"$regex": 
                                '^(design|ux|ui|design|frontend|backend|gaming|game|gamer|web|develop|software)',
                                "$options" :'i'}})
```

```python
companies_tag = comp.find({'tag_list':
                               {"$regex": 
                                '^(design|ux|ui|design|frontend|backend|gaming|game|gamer|web|develop|software)',
                                "$options" :'i'}})
```

```python
companies_desc = comp.find({'description':
                               {"$regex": 
                                '^(design|ux|ui|design|frontend|backend|gaming|game|gamer|web|develop|software)',
                                "$options" :'i'}})
```
